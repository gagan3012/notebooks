```python
# Importing the PIL library
from PIL import Image
from PIL import ImageDraw
from PIL import ImageFont
import re
from glob import glob
from datasets import load_dataset, Dataset,concatenate_datasets, load_from_disk, DatasetDict
import pandas as pd
from tqdm import tqdm
from shutil import make_archive
import os, shutil
import random
random.seed(42)

from datasets.utils.logging import disable_progress_bar
disable_progress_bar()

fonts = [
    "Cairo",
    "Gulzar",
    "Tajawal",
    "Almarai",
    "Amiri",
    "IBM Plex Sans Arabic",
    "Changa",
    "El Messiri",
    "Mada",
    "Readex Pro",
    "Lalezar",
    "Noto Naskh Arabic",
    "Markazi Text",
    "Lemonada",
    "Reem Kufi",
    "Alexandria",
    "Lateef",
    "Blaka",
    "Harmattan",
    "Rakkas",
    "Mirza",
    "Aref Ruqaa",
    "Marhey",
    "Vazirmatn",
    "Noto Nastaliq Urdu",
    "Jomhuria",
    "Kufam",
    "Katibeh",
    "Cairo Play",
    "Baloo Bhaijaan 2",
    "Amiri Quran",
    "Scheherazade New",
    "Alkalami",
    "Blaka Hollow",
    "Qahiri",
    'Vibes',
    'Blaka Ink',
    'Reem Kufi Fun',
    'Reem Kufi Ink',
    'Noto Sans Arabic',
    'Noto Kufi Arabic',
    "Cairo Play",
    "Aref Ruqaa Ink",
]

def find_font_size(text, font, image, target_width_ratio):
    tested_font_size = 100
    tested_font = ImageFont.truetype(font, tested_font_size)
    observed_width, observed_height = get_text_size(text, image, tested_font)
    estimated_font_size = tested_font_size / (observed_width / image.width) * target_width_ratio
    return round(estimated_font_size)

def get_text_size(text, image, font):
    im = Image.new('RGB', (image.width, image.height))
    draw = ImageDraw.Draw(im)
    return draw.textsize(text, font,language="ar")


def check_space(string):

    # counter
    count = 0

    # loop for search each index
    for i in range(0, len(string)):

        # Check each char
        # is blank or not
        if string[i] == " ":
            count += 1

    return count

def delete(folder):
    for filename in os.listdir(folder):
      file_path = os.path.join(folder, filename)
      try:
          if os.path.isfile(file_path) or os.path.islink(file_path):
              os.unlink(file_path)
          elif os.path.isdir(file_path):
              shutil.rmtree(file_path)
      except Exception as e:
          print('Failed to delete %s. Reason: %s' % (file_path, e))

    os.rmdir(folder)


def generate(idx,image, text):
    def flatten(l):
        return [item for sublist in l for item in sublist]

    fonts_arabic = flatten(
        [glob(f"""/content/fonts/ofl/{x.replace(" ","").lower()}/*.ttf""") for x in fonts])

    data = []

    fonts_arabic = random.sample(fonts_arabic,5)
    
    w, h = image.size

    # data.append((image, text))

    os.makedirs(f"/content/NewOnlineKhattv2/", exist_ok=True)
    if os.path.exists(f"/content/NewOnlineKhattv2/Images_{idx}/"):
      delete(f"/content/NewOnlineKhattv2/Images_{idx}/")
    os.makedirs(f"/content/NewOnlineKhattv2/Images_{idx}/", exist_ok=True)

    if check_space(text) > 100:
      text = text.replace(
          "                                                                                                                                                   ", "\n")

    image.save(f"/content/NewOnlineKhattv2/Images_{idx}/original_{idx}.jpeg")

    image_save=f"original_{idx}.jpeg"

    data.append((image_save, text))

    for font in fonts_arabic:
        img = Image.new("RGB", (w, h), (255, 255, 255))
        # print(font)
        # Call draw Method to add 2D graphics in an image
        I1 = ImageDraw.Draw(img)

        # Custom font style and font size

        font_size = find_font_size(text,font,image,0.99)
        myFont = ImageFont.truetype(font, font_size)

        W, H = I1.textsize(text,font=myFont,language="ar")

        I1.text((-(W-w)/2, -(H-h)/2), text, font=myFont, fill=(0, 0, 0), align="center",language="ar")

        font_name = font.split("/")[-1].split(".")[0]

        image_name = f"{font_name}_{idx}.jpeg"

        data.append((image_name, text))

        img.save(f"/content/NewOnlineKhattv2/Images_{idx}/{font_name}_{idx}.jpeg")
    
    df = pd.DataFrame(data, columns=["file_name", "text"])
    df.to_csv(f"/content/NewOnlineKhattv2/Images_{idx}/metadata.csv",index=False)

    with HiddenPrints():
        dataset = load_dataset("imagefolder", data_dir=f"/content/NewOnlineKhattv2/Images_{idx}/",split="train")

    dataset.save_to_disk(f"/content/Images_{idx}")

    delete(f"/content/NewOnlineKhattv2/Images_{idx}/")
```

```python
def run(dataset):
    dfs=[]
    for i in tqdm(range(len(dataset['train']))[:1000]):
      generate(i,dataset['train'][i]['image'],dataset['train'][i]['text'])
      dfs.append(load_from_disk(f"/content/Images_{i}",keep_in_memory=True))
      delete(f"/content/Images_{i}")

    data = DatasetDict({
        "train":concatenate_datasets(dfs),
        "validation":dataset['dev'],
        "test":dataset['test'],
    })
    data.push_to_hub(f"gagan3012/NewArOCRDatasetv3")

    return dfs
```
